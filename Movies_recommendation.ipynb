```python
df = df.drop(['budget', 'homepage', 'original_language','original_title', 'overview', 'popularity', 'production_countries', 
         'release_date', 'revenue', 'runtime','spoken_languages', 'status', 'tagline', 'vote_count','movie_id'], axis = 1)
```

```python
def convert(obj):
    l= []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
df['genres'] = df['genres'].apply(convert)
```

```python
df['keywords'] = df['keywords'].apply(convert)
```

```python
def convert3(obj):
    l= []
    c = 0
    for i in ast.literal_eval(obj):
        if c !=3:
            l.append(i['name'])
            c+=1
        else:
            break
    return l
```

```python
df['cast'] = df['cast'].apply(convert3)
```

```python
df.iloc[0].crew
```

```python
def convert2(obj):
    l= []
    for i in ast.literal_eval(obj):
        if i['job'] =='Director':
            l.append(i['name'])
        
    return l
```

```python
df['crew'] = df['crew'].apply(convert2)
```

```python
df['vote_average'] = df['vote_average'].astype(str)
```

```python
df.iloc[0].production_companies
```

```python
df['production_companies'] = df['production_companies'].apply(convert)
```

```python
df.head(5)
```

```python
df['genres'] = df['genres'].apply(lambda x:[i.replace(' ','') for i in x])
df['keywords'] = df['keywords'].apply(lambda x:[i.replace(' ','') for i in x])
df['production_companies'] = df['production_companies'].apply(lambda x:[i.replace(' ','') for i in x])
df['cast'] = df['cast'].apply(lambda x:[i.replace(' ','') for i in x])
df['crew'] = df['crew'].apply(lambda x:[i.replace(' ','') for i in x])
df['vote_average'] = df['vote_average'].astype(str)
df['vote_average'] = df['vote_average'].apply(lambda x:[x])
```

```python
df.head(5)
```

```python
df['tags'] = df['genres'] + df['keywords'] + df['production_companies'] + df['cast'] + df['crew'] + df['vote_average']
```

```python
df['tags']
```

```python
df['tags'] = df['tags'].apply(lambda x:' '.join(x))
```

```python
df['tags'] = df['tags'].apply(lambda x:x.lower())
```

```python
df_new = df.drop(['genres', 'vote_average', 'keywords','production_companies','cast', 'crew'], axis = 1)
```

```python
df_new.head()
```

```python
from nltk.stem.porter import PorterStemmer 
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
df_new['tags'] = df_new['tags'].apply(stem)
```

```python
df_new['tags'][2]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words ='english')
```

```python
vactors = cv.fit_transform(df_new['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
cs_score = cs(vactors)
```

```python
def recommend(movie):
    movie_index = df_new[df['title'] == movie].index[0]
    distances = cs_score[movie_index]
    movies_list = sorted(list(enumerate(distances)), reverse= True, key = lambda x:x[1])[1:7]
    
    for i in movies_list:
        
        print(df_new.iloc[i[0]].title)
    
```

```python
recommend('Superman Returns')
```
